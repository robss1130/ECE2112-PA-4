## Problem 1 ##


```python
#Filter the students from Luzon with Instrumentation as their track
Instru = df[(df['Track'] == 'Instrumentation') & (df['Hometown'] == 'Luzon') & (df['Electronics'] > 70)]
```

```python
#Pick out the wanted columns to display the same output as the given output
Instru = Instru[['Name', 'GEAS', 'Electronics']]
```

```python
#Filter out the female students with Mindanao as their hometown
Mindy = df[(df['Hometown'] == 'Mindanao') & (df['Gender'] == 'Female')].copy()
```

```python
#Calculate the average of their scores
Mindy.loc[:, 'Average'] = Mindy[['Math', 'GEAS', 'Electronics', 'Communication']].mean(axis=1)
#Include only the students with an average score that is greater or equal to 55
Mindy = Mindy[Mindy['Average'] >= 55]
```

```python
#Pick out wanted columns for desired output
Mindy = Mindy[['Name', 'Track', 'Electronics', 'Average']]
```

## 2. Create a visualization to show different factors of the average grade. ##

```python
#Use boxplot as the visualization technique to display the factors to the avg grade
df['Average'] = df[['Math', 'GEAS', 'Electronics', 'Communication']].mean(axis=1)
print('Features that contributes to average grades')
sns.boxplot(x='Track', y='Average', hue='Gender', data=df)
plt.show()
```

```python
print('Average Scores by Hometown')
plt.figure(figsize=(10, 6))
sns.boxplot(x='Hometown', y='Average', data=df)
plt.show()
```
